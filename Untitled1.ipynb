```python
import requests

import pandas as pd

url1 = "https://ghoapi.azureedge.net/api/NCD_BMI_30C"
df1 = pd.DataFrame(requests.get(url1).json()['value'])
df1['IndicatorCode'] = 'NCD_BMI_30C'
df1['age_group'] = 'Adult'

# Dataset 3
url2 = "https://ghoapi.azureedge.net/api/NCD_BMI_PLUS2C"
df2 = pd.DataFrame(requests.get(url2).json()['value'])
df2['IndicatorCode'] = 'NCD_BMI_PLUS2C'
df2['age_group'] = 'Child'


# Dataset 2
url3 = "https://ghoapi.azureedge.net/api/NCD_BMI_18C"
df3 = pd.DataFrame(requests.get(url3).json()['value'])
df3['IndicatorCode'] = 'NCD_BMI_18C'
df3['age_group'] = 'Adult'


# Dataset 4
url4 = "https://ghoapi.azureedge.net/api/NCD_BMI_MINUS2C"
df4 = pd.DataFrame(requests.get(url4).json()['value'])
df4['IndicatorCode'] = 'NCD_BMI_MINUS2C'
df4['age_group'] = 'Child'
```

```python
#Create Table
cur.execute("""CREATE TABLE Obesity (region VARCHAR(255),
        Gender VARCHAR(10),
        Year INT,
        Lowerbound FLOAT,
        Upperbound FLOAT,
        Mean_Estimate FLOAT,
        Country VARCHAR(255),
        age_group VARCHAR(10),
        CI_WIDTH FLOAT,
        Obesity_level VARCHAR(255)
        
    )
""")



cur.execute("""CREATE TABLE Malnutrition (region VARCHAR(255),
        Gender VARCHAR(10),
        Year INT,
        Lowerbound FLOAT,
        Upperbound FLOAT,
        Mean_Estimate FLOAT,
        Country VARCHAR(255),
        age_group VARCHAR(10),
        CI_WIDTH FLOAT,
        Malnutrition_level VARCHAR(255)
        
    )
""")
```

```python
# insert the data in Obesity

insert_query = """
INSERT INTO Obesity (
    Region, Gender, Year, Lowerbound,
    Upperbound, Mean_Estimate, Country,
    age_group, CI_WIDTH, Obesity_level
)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

for idx, row in df_obesity.iterrows():
    data_tuple = (
        row['Region'],
        row['Gender'],
        int(row['Year']),
        float(row['Lowerbound']),
        float(row['Upperbound']),
        float(row['Mean_Estimate']),
        row['Country'],
        row['age_group'],
        float(row['CI_WIDTH']),
        row['Obesity_level']
    )
    cur.execute(insert_query, data_tuple)

myconnection.commit()
```

```python
# OBESITY TABLE QUERIES

# 1.Top 5 regions with highest average obesity levels in 2022
import pandas as pd

obesity_data = pd.read_csv('obesity_data.csv')
top_regions_2022 = (
    obesity_data[obesity_data['Year'] == 2022]
    .groupby('Region')['Mean_Estimate']
    .mean()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)
print("Top 5 Regions (2022):\n", top_regions_2022)
```

```python
# 2. Top 5 countries with highest obesity estimates (latest year)
latest_year = obesity_data['Year'].max()
top_countries_latest = (
    obesity_data[obesity_data['Year'] == latest_year]
    .groupby('Country')['Mean_Estimate']
    .max()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)
print("Top 5 Countries (Latest Year):\n", top_countries_latest)
```

```python
# 3. Obesity trend in India over the years (Mean_estimate)
india_trend = (
    obesity_data[obesity_data['Country'] == 'India']
    .groupby('Year')['Mean_Estimate']
    .mean()
    .reset_index()
)
print("Obesity Trend in India:\n", india_trend)
```

```python
# 4. Average obesity by gender
avg_obesity_gender = (
    obesity_data.groupby('Gender')['Mean_Estimate']
    .mean()
    .reset_index()
)
print("Average Obesity by Gender:\n", avg_obesity_gender)
```

```python
 # 5. Country count by obesity level category and age group
bins = [0, 10, 20, 30, 100]
labels = ['Low', 'Moderate', 'High', 'Very High']
obesity_data['Obesity_Level_Category'] = pd.cut(obesity_data['Mean_Estimate'], bins=bins, labels=labels)

country_count_by_level_age = (
    obesity_data.groupby(['Obesity_Level_Category', 'age_group'])['Country']
    .nunique()
    .reset_index()
    .rename(columns={'Country': 'Country_Count'})
)
print("Country Count by Obesity Level and Age Group:\n", country_count_by_level_age)
```

```python
# 6. Top 5 least reliable countries (highest average CI_Width) & top 5 most consistent (lowest avg CI_Width)
obesity_data['CI_Width'] = obesity_data['Upperbound'] - obesity_data['Lowerbound']

avg_CI_by_country = (
    obesity_data.groupby('Country')['CI_Width']
    .mean()
    .reset_index()
)

least_reliable = avg_CI_by_country.sort_values('CI_Width', ascending=False).head(5)
most_consistent = avg_CI_by_country.sort_values('CI_Width', ascending=True).head(5)

print("Top 5 Least Reliable Countries:\n", least_reliable)
print("Top 5 Most Consistent Countries:\n", most_consistent)
```

```python
# 7. Average obesity by age group
avg_obesity_age_group = (
    obesity_data.groupby('age_group')['Mean_Estimate']
    .mean()
    .reset_index()
)
print("Average Obesity by Age Group:\n", avg_obesity_age_group)
```

```python
# 8. Top 10 countries with consistent low obesity (low avg + low CI) over years
country_stats = (
    obesity_data.groupby('Country')
    .agg({'Mean_Estimate': 'mean', 'CI_Width': 'mean'})
    .reset_index()
)

consistent_low = country_stats[(country_stats['Mean_Estimate'] < 10) & (country_stats['CI_Width'] < 5)]
top_10_consistent_low = consistent_low.sort_values(['Mean_Estimate', 'CI_Width']).head(10)
print("Top 10 Countries with Consistent Low Obesity:\n", top_10_consistent_low)
```

```python
# 9. Countries where female obesity exceeds male by large margin (same year)
pivot_gender = obesity_data.pivot_table(
    index=['Country', 'Year'],
    columns='Gender',
    values='Mean_Estimate'
).reset_index()

pivot_gender['Female_Male_Diff'] = pivot_gender['Female'] - pivot_gender['Male']
large_diff = pivot_gender[pivot_gender['Female_Male_Diff'] > 5]  # Adjust threshold as needed
print("Countries where Female Obesity Exceeds Male by Large Margin:\n", large_diff)
```

```python
# 10. Global average obesity percentage per year
global_avg_per_year = (
    obesity_data.groupby('Year')['Mean_Estimate']
    .mean()
    .reset_index()
)
print("Global Average Obesity % per Year:\n", global_avg_per_year)
```

```python
import pandas as pd

# Load CSVs
obesity_data = pd.read_csv('obesity_data.csv')
malnutrition_data = pd.read_csv('malnutrition_data.csv')

# Convert column names to lowercase
obesity_data.columns = obesity_data.columns.str.lower()
malnutrition_data.columns = malnutrition_data.columns.str.lower()
```

```python
#combine table queries
# 1. Any 5 countries comparision

import pandas as pd
obesity = pd.read_csv('obesity_data.csv')
malnutrition = pd.read_csv('malnutrition_data.csv')



combined = pd.concat([obesity, malnutrition])
selected_countries = ['India', 'USA', 'Brazil', 'Nigeria', 'Germany']


query1 = (
    combined[combined['country'].isin(selected_countries)]
    .groupby(['country', 'condition'])['mean_estimate']
    .mean()
    .reset_index()
    .rename(columns={'mean_estimate': 'avg_estimate'})
)

print("\n1. Obesity vs Malnutrition Comparison (Any 5 Countries):")
print(query1)
```

```python
#2. Gender-Wise 

query2 = (
    combined.groupby(['gender', 'condition'])['mean_estimate']
    .mean()
    .reset_index()
    .rename(columns={'mean_estimate': 'avg_estimate'})
)

print("\n2. Gender-based Disparity in Obesity and Malnutrition:")
print(query2)
```

```python
# 3. Region Wise
query3 = (
    combined[combined['region'].isin(['Africa', 'Americas'])]
    .groupby(['region', 'condition'])['mean_estimate']
    .mean()
    .reset_index()
    .rename(columns={'mean_estimate': 'avg_estimate'})
)

print("\n3. Region-wise Average Estimates (Africa and Americas):")
print(query3)
```

```python
# 4. Countries Max & Min 
import pandas as pd

obesity = pd.read_csv('obesity_data.csv')
malnutrition = pd.read_csv('malnutrition_data.csv')

obesity.columns = obesity.columns.str.lower()
malnutrition.columns = malnutrition.columns.str.lower()


obesity_trend = (
    obesity.groupby('country')['mean_estimate']
    .agg(['min', 'max'])
    .reset_index()
    .rename(columns={'min': 'ob_min', 'max': 'ob_max'})
)
obesity_trend['ob_change'] = obesity_trend['ob_max'] - obesity_trend['ob_min']

# Malnutrition trend
malnutrition_trend = (
    malnutrition.groupby('country')['mean_estimate']
    .agg(['min', 'max'])
    .reset_index()
    .rename(columns={'min': 'mal_min', 'max': 'mal_max'})
)
malnutrition_trend['mal_change'] = malnutrition_trend['mal_max'] - malnutrition_trend['mal_min']
trend_compare = pd.merge(obesity_trend, malnutrition_trend, on='country')
query4 = trend_compare[
    (trend_compare['ob_change'] > 0) & (trend_compare['mal_change'] < 0)
][['country', 'ob_change', 'mal_change']]

print("\n4. Countries with Obesity ↑ and Malnutrition ↓:")
print(query4)
```

```python
# 5. Age Wise 

query5 = (
    combined.groupby(['age_group', 'year', 'condition'])['mean_estimate']
    .mean()
    .reset_index()
    .rename(columns={'mean_estimate': 'avg_estimate'})
    .sort_values(by=['age_group', 'year'])
)

print("\n5. Age-wise Trend Analysis:")
print(query5)
```
